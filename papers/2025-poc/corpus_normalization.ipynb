```python
import re
import string
from datetime import datetime

start = datetime.now()
print(f'Starting corpus normalization: {start}')

#source_file = 'monolingual.hi'
#target_file = 'monolingual-04-08-2024.hi'
source_file = 'test.txt'
target_file = 'test2.txt'
subsitution_count = 0

pattern = re.compile(r'(?<!\w)(' + '|'.join(re.escape(key) for key in replace_table_regex.keys()) + r')(?!\w)')
#pattern = re.compile(r'\b(' + '|'.join(re.escape(key) for key in replace_table_regex.keys()) + r')\b')


def repl(m):
    return replace_table_regex[m.group(0)]

sf = open(source_file, 'r')
tf = open(target_file, 'w')
while True:
    filedata = sf.readline()
    filedata = filedata.replace('।', ' ')
    remove_punctuation_marks = str.maketrans(string.punctuation, ' '*len(string.punctuation))
    filedata = filedata.translate(remove_punctuation_marks)
    filedata = re.sub(' +', ' ', filedata)
    filedata, temp = pattern.subn(lambda x: replace_table_regex[x.group()], filedata)
    subsitution_count += temp
    
    tf.write(filedata)
    if not filedata:
        break
    
sf.close()
tf.close()
end = datetime.now()
print(f'Number of performed substitutions: {subsitution_count}')
print(f'Ending corpus normalization: {end - start}')
```
